```python
def del_cost():
    return 1

def ins_cost():
    return 1

def sub_cost(c1, c2):
    if c1 == c2:
        return 0
    else:
        return 2


def min_edit_distance_version1(source, target, do_print_chart=False):
    n = len(source)
    m = len(target)
    Distance = [[0] * (m + 1) for _ in range(n + 1)]
    Distance[0][0]=0

    # Initialiser la première ligne et la première colonne avec des indices
    for i in range(1,n + 1):
        Distance[i][0]=Distance[i-1][0]+del_cost()
    for j in range(1,m + 1):
        Distance[0][j]=Distance[0][j-1]+ins_cost()

    # La relation de recurrence:
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            Distance[i][j] = min(
                Distance[i - 1][j] + del_cost(),
                Distance[i][j - 1] + ins_cost(),
                Distance[i - 1][j - 1] + sub_cost(source[i - 1], target[j - 1])
            )

    if do_print_chart:
        for row in Distance:
            print(row)
    return Distance[n][m]


if __name__ == '__main__':
    import sys

    if len(sys.argv) == 3:
        w1 = sys.argv[1]
        w2 = sys.argv[2]
    w1 = 'intention'
    w2 = 'execution'
print(f'edit distance between  intention and execution is: {min_edit_distance_version1("intention", "execution",do_print_chart=True)}')

```

```python
def min_edit_distance_version2(source, target, do_print_chart=False, do_print_backtrace=False, do_print_alignment=False):
    n = len(source)
    m = len(target)
    Distance = [[0] * (m + 1) for _ in range(n + 1)]
    Backtrace = [[None] * (m + 1) for _ in range(n + 1)]
    Distance[0][0] = 0

    # Initialiser la première ligne et la première colonne avec des indices
    for i in range(1, n + 1):
        Distance[i][0] = Distance[i - 1][0] + del_cost()
        Backtrace[i][0] = 'D'
    for j in range(1, m + 1):
        Distance[0][j] = Distance[0][j - 1] + ins_cost()
        Backtrace[0][j] = 'I'

    # La relation de recurrence avec backtrace:
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            deletion = Distance[i - 1][j] + del_cost()
            insertion = Distance[i][j - 1] + ins_cost()
            substitution = Distance[i - 1][j - 1] + sub_cost(source[i - 1], target[j - 1])

            min_cost = min(deletion, insertion, substitution)
            Distance[i][j] = min_cost

            # Mettre a jour backtrace
            if deletion==insertion==substitution:
                Backtrace[i][j] = 'All' # All for  all operations
            elif min_cost == deletion:
                Backtrace[i][j] = 'D'  # D for suppression
            elif min_cost == insertion:
                Backtrace[i][j] = 'I'  # I for insertion
            else:
                Backtrace[i][j] = 'S'  # S for substitution

    if do_print_chart:
        for row in Distance:
            print(row)

    if do_print_backtrace:
        # Afficher backtrace matrix
        for row in Backtrace:
            print(row)

    if do_print_alignment:
        # Afficher alignment
        align_source = ''
        align_target = ''
        i, j = n, m
        while i > 0 or j > 0:
            if i > 0 and j > 0 and Backtrace[i][j] == 'S':
                align_source = source[i - 1] + align_source
                align_target = target[j - 1] + align_target
                i -= 1
                j -= 1
            elif i > 0 and (j == 0 or Distance[i][j] == Distance[i - 1][j] + del_cost()):
                align_source = source[i - 1] + align_source
                align_target = '-' + align_target
                i -= 1
            else:
                align_source = '-' + align_source
                align_target = target[j - 1] + align_target
                j -= 1
        print("Alignment:")
        print(align_source)
        print(align_target)

    return Distance[n][m], Backtrace


if __name__ == '__main__':
    import sys


    w1 = 'intention'
    w2 = 'execution'

    edit_distance, backtrace = min_edit_distance_version2(w1, w2, do_print_chart=True, do_print_backtrace=True, do_print_alignment=True)
    print(f'Edit distance between "{w1}" and "{w2}" is: {edit_distance}')
```

```python
def load_keyboard_structure(filename):
    with open(filename, 'r') as file:
        keyboard_structure = eval(file.read())
    return keyboard_structure

# Charger le fichier
keyboard_structure = load_keyboard_structure('/content/drive/MyDrive/NLP/pa_files/qwerty_graph.txt')

# Fonction de coût de substitution basée sur la proximité des touches sur le clavier QWERTY
def sub_cost_qwerty(c1, c2):
    if c1 == c2:
        return 0  # Pas de substitution nécessaire pour les mêmes caractères

    # Vérifier si les caractères sont dans le dictionnaire de touches adjacentes
    if c1 in keyboard_structure and c2 in keyboard_structure[c1].values():
        return 1  # Caractères adjacents, coût de substitution plus faible

    return 2  # Coût de substitution par défaut pour les caractères non adjacents

# Exemple d'utilisation
print(sub_cost_qwerty('a', 's'))
```

**AUTRE TECHNIQUE DE NOTRE CHOIX:**
Utilisation de TF-IDF pour la correction orthographique :
 - TF-IDF est une technique de pondération souvent utilisée en traitement automatique du langage naturel et en récupération d'information.
 - Pour chaque mot dans le dictionnaire, nous calculons un score TF-IDF, qui mesure l'importance relative du mot dans le contexte de tous les mots disponibles.
- TF (Term Frequency) mesure le nombre de fois qu'un mot apparaît dans un document (dans notre cas, un mot dans le dictionnaire).
- IDF (Inverse Document Frequency) mesure l'inverse de la fréquence du mot dans l'ensemble des documents (dans notre cas, l'ensemble des mots du dictionnaire).
 - Le produit de TF et IDF donne le score TF-IDF, qui est utilisé pour représenter la signification relative de chaque mot.
 - En utilisant la similarité cosinus entre les vecteurs TF-IDF des mots, nous pouvons mesurer la similarité entre deux mots et les utiliser pour la correction orthographique.
 - En comparant le mot mal orthographié avec les mots du dictionnaire en utilisant TF-IDF, nous pouvons trouver les mots du dictionnaire les plus similaires et suggérer des corrections potentielles.
 - Cette approche nous permet de tenir compte du contexte global des mots dans le dictionnaire et d'identifier les mots les plus pertinents pour la correction orthographique.


```python
import soundex
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def correct_spelling(word, dictionary, k=10):
    """
    Correct the spelling of a word using a given dictionary of words.
    """
    # Filtrer le dictionnaire pour qu'il ne contienne que les mots commençant par la même lettre que le mot donné
    dictionary_filtered = [w for w in dictionary if w[0] == word[0]]

    threshold = 1 # Ajuster threshold as needed
    word_length = len(word)
    dictionary_filtered = [w for w in dictionary_filtered if abs(len(w) - word_length) <= threshold]

    # Calculer le code Soundex pour le mot d'entrée
    soundex_code = soundex.Soundex().soundex(word)

    # Filtrer davantage pour ne contenir que les mots avec le même code Soundex
    dictionary_filtered = [w for w in dictionary_filtered if soundex.Soundex().soundex(w) == soundex_code]

    # Utiliser TF-IDF pour la vectorisation des mots
    all_words = [word] + dictionary_filtered
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(all_words)
    # Assurez-vous que le dictionnaire final_tokens_english_dictionary contient des mots avant d'appeler la fonction correct_spelling.
    # Si le dictionnaire est vide, une erreur peut survenir lors de l'appel de la fonction correct_spelling en raison de l'absence de mots à comparer.
    if dictionary_filtered!=[]:
           # Calculer la similarité cosinus entre le mot d'entrée et les autres mots du dictionnaire
           cosine_similarities = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()

           # Trier les indices des mots par ordre décroissant de similarité cosinus
           similar_indices = cosine_similarities.argsort()[::-1]

           # Sélectionner les k mots les plus similaires
           closest_words = [dictionary_filtered[i] for i in similar_indices[:k]]

           return closest_words
    else:
       return []
# Exemple d'utilisation pour le dictionnaire anglais
word_to_correct = "ater"
corrections = correct_spelling(word_to_correct, final_tokens_english_dictionary)
print(f"Corrections possibles pour '{word_to_correct}': {corrections}")
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

def process_data(corpus_file):
    """
    Preprocesses the text data from the given corpus file and returns only the words.
    """
    with open(corpus_file, 'r', encoding='utf-8') as file:
        text = file.read()

        # Convertir le texte en minuscules
        text = text.lower()

        # Tokeniser le texte en utilisant NLTK et exclure la ponctuation
        words = [word for word in word_tokenize(text) if word.isalpha()]

        return words
```

##### Qst3:

Nous allons implimenter le modèle de langue en calculant la probabilité de chaque mot à partir de fichier big.txt, nous pouvons utiliser la fonction process_data pour prétraiter le texte et ensuite compter le nombre d'occurrences de chaque mot dans le corpus.
Après, nous divisons le nombre d'occurrences de chaque mot par la taille totale du corpus

```python
def build_language_model(corpus_file):

    # Prétraitement du texte du corpus
    words = process_data(corpus_file)

    # Comptage du nombre d'occurrences de chaque mot
    word_counts = {}
    total_words = len(words)
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    # Calcul des probabilités de chaque mot
    language_model = {}
    for word, count in word_counts.items():
        language_model[word] = count / total_words

    return language_model

# Exemple d'utilisation pour construire le modèle de langue
corpus_file ='/content/drive/MyDrive/pa_files/big.txt'
language_model = build_language_model(corpus_file)

# Affichage de la probabilité de quelques mots pour vérification
words_to_check = ['the', 'quick', 'brown', 'fox']
for word in words_to_check:
    print(f"Probability of '{word}': {language_model.get(word, 0)}")
```

Nous allons implimenter les fonctions edits1 (une modification ) et edits2 (2 modifications).

```python
def edits1(s):
    # Retourne l'ensemble de toutes les chaînes pouvant être obtenues avec une modification à effectuer sur la chaîne s.
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    splits = [(s[:i], s[i:]) for i in range(len(s) + 1)]
    deletes = [a + b[1:] for a, b in splits if b]
    replaces = [a + c + b[1:] for a, b in splits for c in alphabet if b]
    inserts = [a + c + b for a, b in splits for c in alphabet]
    return set(deletes + replaces + inserts)

def edits2(s):
    # Retourne l'ensemble de toutes les chaînes pouvant être obtenues avec deux modifications à effectuer sur la chaîne s.

    return set(e2 for e1 in edits1(s) for e2 in edits1(e1))
```

```python
def knownWords(words, vocabulary):
    # Filtre les mots de la liste words qui ne sont pas dans le vocabulaire.
    return [word for word in words if word in vocabulary]
```

##### Qst5:

La fonction suivante retourne la première liste non vide de candidats par ordre de priorité : le mot original, les mots connus à une distance d'édition de un, les mots connus à une distance d'édition de deux, le mot original même s'il n'est pas connu.
  

```python
def candidates(word):
    # Si le mot original est dans le vocabulaire, retournez-le comme seul candidat avec sa probabilité
    if word in language_model:
        return [(word, language_model[word])]

    # Recherchez les mots connus à une distance d'édition de un
    candidates_1 = knownWords(edits1(word), vocabulary)
    candidates_1_with_prob = [(c, language_model.get(c, 0)) for c in candidates_1]
    if candidates_1_with_prob:
        return candidates_1_with_prob

    # Recherchez les mots connus à une distance d'édition de deux
    candidates_2 = knownWords(edits2(word), vocabulary)
    candidates_2_with_prob = [(c, language_model.get(c, 0)) for c in candidates_2]
    if candidates_2_with_prob:
        return candidates_2_with_prob

    # Si aucune liste de candidats n'est trouvée, retournez simplement le mot original avec une probabilité nulle
    return [(word, 0)]

# Exemple d'utilisation
word = "techer"
print("Candidates for 'techer':", candidates(word))

```

##### Qst6:

Implimentons la fonction correction(word, k)

```python
def correction(word, k):
    # Les corrections candidats
    candidates_list = candidates(word)

    # Mettre en ordre les resultats
    sorted_candidates = sorted(candidates_list, key=lambda x: x[1], reverse=True)

    # Retourne les top k corrections probables
    top_k_corrections = sorted_candidates[:k]

    return top_k_corrections
# Exemple d'utilisation
word_to_correct = "helo"
corrections = correction(word_to_correct, 5)
print(f"Corrections possibles pour '{word_to_correct}': {corrections}")
```
